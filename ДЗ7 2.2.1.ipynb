### деление таблицы на две части (по строкам) и собрать с помощью функции pd.concat

```python
long = df.shape
print('количество строк и столбцов',long)
top = int(long[0]/2)
down = int(long[0])
print('первая половина строк от 0 до',top-1)
print('вторая половина строк', top,'до',down)
```

```python
# Первая часть таблицы
df_top = df.iloc [:top]
df_top
```

```python
# Вторая часть таблицы
df_down= df.tail(down-top)
df_down
```

```python
# Объединение таблиц
df_all = pd.concat([df_top,df_down],sort=False,axis=0)
df_all
```

```python
#Определяем интервал для выбросов
q1 = df_all_nan['alcohol'].quantile(0.25)
q3 = df_all_nan['alcohol'].quantile(0.75)
iqr = q3 - q1
lowr_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
print (f'Выбросами будут считаться вся продукция(Вино), процент алкоголя в которых находится за пределами интервала [{lowr_bound},{upper_bound}]')
```

```python
#Выбросы
df_all_nan[(df_all_nan['alcohol']<=lowr_bound)|(df_all_nan['alcohol']>=upper_bound)]
```

```python
#Значение которые попадают в интервал
df_all_nan_ejection=df_all_nan[(df_all_nan['alcohol']>=lowr_bound)&(df_all_nan['alcohol']<=upper_bound)]
df_all_nan_ejection
```

### Нормализация данных

```python
# делаем копию таблицы
df_all_nan_ejection_norm=df_all_nan_ejection.copy()
```

```python
#нормализуем данные в столбце "alcohol" и записываего их в новый сталбец "norm_alcohol" определив индекс столбца "alcohol" 
df_all_nan_ejection_norm.insert((df_all_nan_ejection_norm.columns.get_loc('alcohol')+1),'norm_alcohol',(df_all_nan_ejection['alcohol'] -df_all_nan_ejection['alcohol'].min())/(df_all_nan_ejection['alcohol'].max()-df_all_nan_ejection['alcohol'].min()))
```
