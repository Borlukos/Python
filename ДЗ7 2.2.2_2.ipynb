### Написать функцию и применить ее к данным с помощью apply()

```python
# функция переводит показатель пролина из мг/дм3 в г/литр 
convector = lambda x: x/1000
# результат записываем в новый столбец proline_g/liter
df['proline_g/liter'] = df['proline'].apply(convector)
```

```python
df
```

### Напишите функцию и примените её к данным с помощью .transform

```python
# функция считает на сколько процентов одно число больше другого
# функция = (Число1/Число2) · 100 − 100
def get_proline_percent(df):
     return round((df.mean()/df)*100-100,2)
```

```python
# записываем процент по модулю в столбец proline_percent.
# proline_percent показывает на сколько процентов proline отличается от среднего значения в процентах.
# Если процент положительный (среднее proline < proline) то делаем вывод Контрафакт.
# То есть на процент (proline_percent) данное вино возможно будет Контрафактом.
# Если процент отрицательный (среднее proline > proline) то делаем вывод Норма.
# То есть на процент (proline_percent) данное вино возможно будет Нормальным.
# Вывод: в приставленной таблицы вин по показателю proline возможны контрофактные партии.
df['proline_percent'] = abs(df.groupby('class_name')['proline'].transform(get_proline_percent))
df['proline_result'] = np.where (df.groupby('class_name')['proline'].transform(get_proline_percent) >0, 'Контрафакт', 'Норма')
df
```

### Напишите функцию и примените её к данным с помощью .map

```python
# функция добавляет к переменной символ %
df['proline_percent'] = df['proline_percent'].map(lambda x: str(x) + '%')
```

### Группировать данные по дате и использовать resample для агрегирования в более крупный период.

```python
# генерирую список дат и записываю их в столбец для отдельных class_name, чтобы даты повторялись
df.insert((df.columns.get_loc('class_name')+1),'date',0)
index = 0
idate = 0
values = df['class_name'].unique()
date = pd.date_range('2016-01-15','2024-09-24', periods = 71 ).date
for value in values:
    idate=0
    while value == df.loc[index,'class_name']:
        df.loc[index,'date'] = date[idate]
        index=index+1
        if index==df['class_name'].shape[0]:
            break
        idate=idate+1
# переводим формат из object в datetime64
df['date']= pd.to_datetime(df['date'])
```

```python
#группируем по дате и агрегируем среднее значение alcohol
df_date = df.groupby('date').agg({'alcohol':'mean'})
df_date
```

```python
# Средне значение alcohol в вине по годам.
# Вывод: современная тенденция уменьшать процент alcohol в вине
df_date.resample('YS').mean()
```
