### загрузка таблицы "Wine.csv" c github

```python
df = pd.read_csv (r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Wine.csv")
```

```python
df
```

### Таблица Wine имеет длинный формат это видно по первым двум столбцам "class_label"	"class_name"

```python
df.style.highlight_between(['class_label','class_name'],color="#00FF00")
```

### Преобразуем таблицу из длинного формата в широкий

```python
# создаем копию таблицы
df_long = df.copy()
```

```python
pd.pivot(df_long,index=['class_label','class_name'],columns = 'alcohol', values ='malic_acid')
```

```python
df_long
```

```python
group_df = df_long.groupby('class_name')
```

```python
group_df = df_long.groupby(['class_label', 'class_name']).agg({'alcohol':'mean','malic_acid':'mean'}).reset_index()
```
