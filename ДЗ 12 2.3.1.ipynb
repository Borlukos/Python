```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from scipy import stats
import statistics
from scipy.stats import mannwhitneyu
import statsmodels.api as sm
```

```python
#загрузка данных
df = pd.read_csv (r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Wine.csv")
df.head()
```

# 1.Очистка от выбросов и пропусков

```python
#проверка на пропуски, пропуски отсутствуют.
df.isna().sum()
```

```python
#проверка на выбросы
df.describe()
```

```python
#Определяем интервал для выбросов (специально взял очень маленький интервал)
q1 = df['alcohol'].quantile(0.40)
q3 = df['alcohol'].quantile(0.60)
iqr = q3 - q1
lowr_bound = round((q1 - 1.5*iqr),2)
upper_bound = round((q3 + 1.5*iqr),2)
print (f'Выбросами будут считаться вся продукция(Вино), процент алкоголя в которых находится за пределами интервала [{lowr_bound,2},{upper_bound}]')
```

```python
#Выбросы
df[(df['alcohol']<=lowr_bound)|(df['alcohol']>=upper_bound)]
```

```python
#убарл выбросы
df_ejection=df[(df['alcohol']>=lowr_bound)&(df['alcohol']<=upper_bound)]
# очищенные данные
df_ejection
```

# 2.Для переменных df_Barolo и df_Grignolino найти 95% доверительные интервалы средних значений

```python
#уникальные значения столбца class_name
df_ejection['class_name'].unique()
```

```python
#Среднее значение переменных Barolo и Grignolino
print(df_Barolo.mean(), df_Grignolino.mean())
```

##### Для переменной df_Barolo собранные данные не будут противоречить распределениям случайной величины с математическим ожиданием от 13.42 до 13.64 см с вероятностью 95%

```python
stats.norm.interval(confidence = 1-alpha, loc = df_Grignolino.mean(), scale = np.sqrt(df_Grignolino.var()/len(df_Grignolino)))
```

# 4.Сравнение переменных df_Barolo и df_Grignoli с ненормальным распределением

```python
#Тест Манна-Уитни
statistic, pvalue = stats.mannwhitneyu(df_Barolo,df_Grignolino,alternative ='two-sided')
print('statistic:', statistic, 'p-значение:' ,pvalue)
print(np.where(pvalue < alpha, "Отвергаем нулевую гипотезу: переменные различаются", "Не отвергаем нулевую гипотезу: переменные не различаются"))
```

##### p =1.498 < 5%, нулевая гипотеза, что распределение алкоголя в вине Barolo и Grignolino совпадают, отвергается. Процент алкоголя в разных винах не равный.

### можно сделать предположение, что вино Barolo имеет больше всего контрафакт, из-за высокого показателя отношения алкоголя к пролину

### Создадим новый столбец "proline_rat" равный отношению среднего значения proline по классу к каждому конкретному значения proline. Среднее значение по классу считаем за эталонное значение proline, если значение proline в вине значительно отклоняется от среднего значения(эталонного значения) это может быть признаком подделки.

```python
#функция отношения среднее значение proline по классу к каждому конкретному значения proline
def get_proline_percent(df):
     return round((df.mean()/df),2)
```

```python
# добавляем столбец proline_rat с данными отношениея среднего значения proline по классу к каждому конкретному значения proline
df_alc_pro = df_ejection.copy()
df_alc_pro['proline_rat'] = abs(df_ejection.groupby('class_name')['proline'].transform(get_proline_percent))
```

```python
# строем график отношения alcohol к вычеслинному proline_rat в вине
sns.lmplot(df_alc_pro,x='alcohol', y='proline_rat', hue= 'class_name')
```

### Можно сделать вывод, что наше предположение, что вино Barolo имеет больше всего контрафакт неверно. По графику видно, что больше всего вероятность контрафакт в вине Grignolino, вывод сделал на основание высокого показателя отношения alcohol к proline_rat в вине.

# Корреляция между переменными на основании статистических тестов Пирсона и Спирмана, в нашем случае используется только тест Спирмана, потому что данные не соответствуют нормальному распределению..

##### Тест Спирмана 
##### Нулевая гипотеза (H0): между двумя переменными нет монотонной зависимости
##### коэффициент корреляции, варьируется от -1 до 1:
##### 1: Идеальная положительная монотонная зависимость (по мере увеличения одной переменной другая также увеличивается).
##### -1: Идеальная отрицательная монотонная зависимость (по мере увеличения одной переменной другая уменьшается).
##### 0: Отсутствие монотонной зависимости.

```python
print(stats.spearmanr(df_alc_pro['proline_rat'], df_alc_pro['alcohol']))
print(np.where(stats.spearmanr(df_alc_pro['proline_rat'], df_alc_pro['alcohol'])[1] < alpha, "Отвергаем нулевую гипотезу: между двумя переменными есть монотонная зависимости. ", "Не отвергаем нулевую гипотезу: между двумя переменными нет монотонной зависимости. "))и
```

### В данном случае p-значение (0.279) больше уровня значимости (0.05), что означает, между двумя переменными нет монотонной зависимости.

```python
print(stats.spearmanr(df_alc_pro['proline_rat'], df_alc_pro['alcohol'])[0])
```

### Вывод: Коэффициент корреляции -0.092 указывает на отрицательную зависимость между alcohol и proline_rat.

### Нет достаточных доказательств, утверждать, что существует статистически значимая монотонная зависимость между  alcohol и proline_rat.

```python
X = np.array(df_alc_pro['alcohol'])
Y = np.array(df_alc_pro['proline_rat'])
X = sm.add_constant(X)
model = sm.OLS(Y, X)
results = model.fit()
print(results.summary()) 
```
